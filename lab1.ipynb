```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

f1 = lambda x: x**2 + 5

x1 = np.linspace(-1, 1)
x2 = np.linspace(-6, 6)
x3 = np.linspace(0, 5)
plt.plot(x1, f1(x1))
plt.show()
```

```python
plt.plot(x2, f1(x2))
plt.show()
```

```python
plt.plot(x3, f1(x3))
plt.show()
```
